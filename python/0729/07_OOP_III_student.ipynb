## `super()`

* ìì‹ í´ë˜ìŠ¤ì— ë©”ì„œë“œë¥¼ ì¶”ê°€ë¡œ êµ¬í˜„í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

* ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë‚´ìš©ì„ ì‚¬ìš©í•˜ê³ ì í•  ë•Œ, `super()`ë¥¼ ì‚¬ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

---

**í™œìš©ë²•**


```python
class ChildClass(ParentClass):
    def method(self, arg):
        super().method(arg) 
```

````python
# Person í´ë˜ìŠ¤ì™€ Student í´ë˜ìŠ¤ë¥¼ í•¨ê»˜ ì •ì˜í•´ ë³´ê² ìŠµë‹ˆë‹¤.
````

````python
class Person:
    def __init__(self, name, age, number, email):
        self.name = name
        self.age = age
        self.number = number
        self.email = email 
        
    def greeting(self):
        print(f'ì•ˆë…•, {self.name}')
      
    
class Student(Person):
    def __init__(self, name, age, number, email, student_id):
        self.name = name
        self.age = age
        self.number = number
        self.email = email 
        self.student_id = student_id
        
p1 = Person('í™êµìˆ˜', 200, '0101231234', 'hong@gildong')
s1 = Student('í•™ìƒ', 20, '12312312', 'student@naver.com', '190000')
````

````python
# p1ê³¼ s1 ëª¨ë‘ greeting ë©”ì„œë“œë¥¼ í˜¸ì¶œí•´ë´…ì‹œë‹¤.
````

````python
p1.greeting()
s1.greeting()
````

ìœ„ì˜ ì½”ë“œëŠ” ìƒì†ì„ í–ˆìŒì—ë„ ë¶ˆêµ¬í•˜ê³  ì´ˆê¸°í™”(`__init__`)ì—ì„œ ë™ì¼í•œ ì½”ë“œê°€ ë°˜ë³µë©ë‹ˆë‹¤. 

ì´ˆê¸°í™”ì˜ ì¤‘ë³µì„ `super()` í•¨ìˆ˜ë¥¼ í†µí•´ ì œê±°í•´ë´…ì‹œë‹¤.

````python
class Person:
    def __init__(self, name, age, number, email):
        self.name = name
        self.age = age
        self.number = number
        self.email = email 
        
    def greeting(self):
        print(f'ì•ˆë…•, {self.name}')
        
        
class Student(Person):
    def __init__(self, name, age, number, email, student_id):
        # Person í´ë˜ìŠ¤
        super().__init__(name, age, number, email)
        self.student_id = student_id
        
        
p1 = Person('í™êµìˆ˜', 200, '0101231234', 'hong@gildong')
s1 = Student('í•™ìƒ', 20, '12312312', 'student@naver.com', '190000')

p1.greeting()
s1.greeting()
````

### [ì‹¤ìŠµ] Rectangle & Square

ì•„ë˜ì˜ ì¡°ê±´ì— ë§Œì¡±í•˜ëŠ” í´ë˜ìŠ¤ `Rentangle` ì„ ì‘ì„±í•˜ì„¸ìš”.

---

> Rectangle í´ë˜ìŠ¤ëŠ” ì•„ë˜ì™€ ê°™ì€ ì†ì„±ê³¼ ë©”ì„œë“œë¥¼ ê°–ìŠµë‹ˆë‹¤.
- ì¸ìŠ¤í„´ìŠ¤ ì†ì„±
    - `length`: ê°€ë¡œ ê¸¸ì´
    - `width`: ì„¸ë¡œ ê¸¸ì´
>
>   
- ì¸ìŠ¤í„´ìŠ¤ ë©”ì„œë“œ
    - `area`: ì§ì‚¬ê°í˜•ì˜ ë„“ì´ë¥¼ ë¦¬í„´
    - `perimeter`: ì§ì‚¬ê°í˜•ì˜ ë‘˜ë ˆì˜ ê¸¸ì´ë¥¼ ë¦¬í„´

````python
# Rectangle í´ë˜ìŠ¤ë¥¼ ì •ì˜í•´ë´…ì‹œë‹¤.
````

````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2 * (self.length + self.width)
````

````python
# Rectangle í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ rectangleì„ ê°€ë¡œ ê¸¸ì´ 4, ì„¸ë¡œ ê¸¸ì´ 8ë¡œ ì´ˆê¸°í™”í•˜ê³ 
# area ë©”ì„œë“œì™€ perimeter ë©”ì„œë“œë¥¼ í†µí•´ ë„“ì´ì™€ ë‘˜ë ˆë¥¼ êµ¬í•´ë´…ì‹œë‹¤.
````

````python
rectangle = Rectangle(4, 8)
print(rectangle.area())
print(rectangle.perimeter())
````

* Rectangle í´ë˜ìŠ¤ë¥¼ ìƒì†ë°›ì•„ Sqaure í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ ì£¼ì„¸ìš”.
    * Square í´ë˜ìŠ¤ëŠ” Rectangle í´ë˜ìŠ¤ì—ì„œ ìƒì†ë°›ì€ ì†ì„± ì™¸ ì¶”ê°€ ì†ì„±ì„ ê°€ì§€ê³  ìˆì§€ ì•ŠìŠµë‹ˆë‹¤.

````python
# Square í´ë˜ìŠ¤ë¥¼ ì •ì˜í•´ë´…ì‹œë‹¤.
````

````python
class Square(Rectangle):
    def __init__(self, length):
        super().__init__(length, length)
````

````python
# Square í´ë˜ìŠ¤ì˜ ì¸ìŠ¤í„´ìŠ¤ squareë¥¼ ê¸¸ì´ 4ë¡œ ì´ˆê¸°í™”í•˜ê³ 
# area ë©”ì„œë“œì™€ perimeter ë©”ì„œë“œë¥¼ í†µí•´ ë„“ì´ì™€ ë‘˜ë ˆë¥¼ êµ¬í•´ë´…ì‹œë‹¤.
````

````python
square = Square(4)
print(square.area(), square.perimeter())
````

# ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©
> Method Overriding(ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©): ìì‹ í´ë˜ìŠ¤ì—ì„œ ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ë©”ì„œë“œë¥¼ ì¬ì •ì˜í•˜ëŠ” ê²ƒ

* ìƒì† ë°›ì€ ë©”ì„œë“œë¥¼ `ì¬ì •ì˜`í•  ìˆ˜ë„ ìˆìŠµë‹ˆë‹¤. 
* ìƒì† ë°›ì€ í´ë˜ìŠ¤ì—ì„œ **ê°™ì€ ì´ë¦„ì˜ ë©”ì„œë“œ**ë¡œ ë®ì–´ì”ë‹ˆë‹¤.

````python
class Person:
    def __init__(self, name, age, number, email):
        self.name = name
        self.age = age
        self.number = number
        self.email = email 
        
    def talk(self):
        print(f'ì•ˆë…•, {self.name}')
````

````python
# ìœ„ì˜ Person í´ë˜ìŠ¤ë¥¼ ìƒì† ë°›ì•„ êµ°ì¸ë‹µê²Œ ë§í•˜ëŠ” Soldier í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ ë³´ê² ìŠµë‹ˆë‹¤.
# talk ë©”ì„œë“œë¥¼ ì¬ì •ì˜(override) í•˜ê² ìŠµë‹ˆë‹¤.
````

````python
class Soldier(Person):
    def __init__(self, name, age, number, email, army):
        super().__init__(name, age, number, email)
        self.army = army
        
    # method overriding    
    def talk(self):
        print(f'ì¶©ì„±! {self.army} {self.name}')
````

````python
p = Person('ì¼ë°˜ì¸', 10, '010123', '1banin@gmail.com')
p.talk()
````

````python
s = Soldier('êµ³ê±´ì´', 25, '0101234', 'soldier@roka.kr', 'í•˜ì‚¬')
s.talk()
````


````python
# MyClass í´ë˜ìŠ¤ë¥¼ ì •ì˜í•´ ë‘ì—ˆìŠµë‹ˆë‹¤.
class MyClass:
    def instance_method(self):
        return self
    
    @classmethod
    def class_method(cls):
        return cls
    
    @staticmethod
    def static_method(arg):
        return arg
````

### [ì½”ë“œì˜ˆì‹œ] Puppy

- Puppy í´ë˜ìŠ¤ì˜ ì†ì„±ì— ì ‘ê·¼í•˜ëŠ” í´ë˜ìŠ¤ ë©”ì„œë“œë¥¼ ìƒì„±í•´ ë´…ì‹œë‹¤.
- í´ë˜ìŠ¤ ë³€ìˆ˜ `population`ë¥¼ í†µí•´ ê°•ì•„ì§€ê°€ ìƒê¸¸ ë•Œë§ˆë‹¤ ì¦ê°€ ì‹œí‚¤ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤.
- ê°•ì•„ì§€ë“¤ì€ ê°ìì˜ ì´ë¦„(`name`)ê³¼ ì¢…(`breed`)ì„ ê°–ê³  ìˆìŠµë‹ˆë‹¤.
- `bark()` ë©”ì„œë“œë¥¼ í˜¸ì¶œí•˜ë©´ ì§–ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

````python
class Puppy:
    population = 0
    
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
        Puppy.population += 1
        
    def __del__(self):
        Puppy.population -= 1
    
    def bark(self):
        print(f'ì™ˆì™ˆ! ë‚˜ëŠ” {self.name}, {self.breed}(ì´)ì•¼')
    
    @classmethod
    def get_population(cls):
        print(f'í˜„ì¬ ê°•ì•„ì§€ ë§ˆë¦¬ìˆ˜: {cls.population}')
        return cls.population
````

````python
# ìœ„ì˜ Puppy í´ë˜ìŠ¤ë¥¼ ì²˜ìŒë¶€í„° ì •ì˜í•´ë´…ì‹œë‹¤.
# ì¶”ê°€ë¡œ ìŠ¤íƒœí‹± ë©”ì„œë“œ infoë¥¼ ì •ì˜í•˜ê³  í˜¸ì¶œí•´ë´…ì‹œë‹¤.
# info ë©”ì„œë“œëŠ” 'ì´ê²ƒì€ Puppy í´ë˜ìŠ¤ì…ë‹ˆë‹¤!' ë¼ëŠ” ë¬¸ìì—´ì„ return í•©ë‹ˆë‹¤.
````

````python
class Puppy:
    population = 0
    
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
        Puppy.population += 1
        
    def __del__(self):
        Puppy.population -= 1
    
    def bark(self):
        print(f'ì™ˆì™ˆ! ë‚˜ëŠ” {self.name}, {self.breed}(ì´)ì•¼')
    
    @classmethod
    def get_population(cls):
        print(f'í˜„ì¬ ê°•ì•„ì§€ ë§ˆë¦¬ìˆ˜: {cls.population}')
        return cls.population
    
    @staticmethod
    def info():
        return 'ì´ê²ƒì€ Puppy í´ë˜ìŠ¤ì…ë‹ˆë‹¤!'
````

````python
d = Puppy('ì´ˆì½”', 'í‘¸ë“¤')

# instance method
d.bark()

# static method
print(d.info(), Puppy.info())

# class method
print(Puppy.get_population())
````

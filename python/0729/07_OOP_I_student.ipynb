## ì‹¤ìŠµ

> ë§¤ì§ë©”ì„œë“œë¥¼ í™œìš©í•˜ì—¬ ì¸ìŠ¤í„´ìŠ¤ê°„ì˜ ë¹„êµì—°ì‚°(`>`, `==`)ì´ ê°€ëŠ¥í•˜ë„ë¡ ë§¤ì§ë©”ì„œë“œë¥¼ ì •ì˜í•´ë´…ì‹œë‹¤.
>
> `Person` í´ë˜ìŠ¤ë¥¼ ì •ì˜í•©ë‹ˆë‹¤. ì¸ìŠ¤í„´ìŠ¤ ì†ì„±ì€ `name` ê³¼ `age`ë¥¼ ê°€ì§€ë©°, ì¸ìŠ¤í„´ìŠ¤ê°„ `age` ë¹„êµê°€ ê°€ëŠ¥í•´ì•¼í•©ë‹ˆë‹¤.
>
> `__gt__`ë©”ì„œë“œì™€ `__eq__`ë©”ì„œë“œë¥¼ í™œìš©í•©ë‹ˆë‹¤.

````python
# ì‹¤ìŠµ ğŸ’œ
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __gt__(self, other):
        return self.age > other.age
    def __eq__(self, other):
        return self.age == other.age
        
````

````python
# ì•„ë˜ ì½”ë“œê°€ ì •ìƒ ë™ì‘í•˜ë„ë¡ í•´ë´…ì‹œë‹¤.
p1 = Person('1', 1)
p2 = Person('2', 2)
p3 = Person('3', 1)

print(p1 > p2)
print(p1 == p3)
print(p1 == p2)
````
